##### Problem 6.2
Calculate average annual percentage return and volatility of all 30 stocks over a theoretical one year period

```python
from pandas import DataFrame
Thirty_data = DataFrame()
Thirty_data['Annual Returns']= stock_data.pct_change().mean()*252
Thirty_data['Volatility']=stock_data.pct_change().std()*np.sqrt(252)
Thirty_data
```

##### Problem 6.3 
Cluster the 30 stocks according to their mean annual Volatilities and Returns using K-means clustering. Identify the optimum number of clusters using the Elbow curve method


```python
from sklearn.cluster import KMeans
from sklearn import metrics
import matplotlib.pyplot as plt
```

```python
## 1st we need to standardise the data (evn though Not required as our data seems already standardised)
from sklearn.preprocessing import StandardScaler

scale = StandardScaler()
X= scale.fit_transform(Thirty_data)
```

```python
## Finding the optimum k value
Sum_of_squared_distances = [] ### Creating an empty array
K = range(1,15)
for k in K:
    km = KMeans(n_clusters=k,random_state=1234) ## random_state is fixed with a seed value for uniformity in results
    km = km.fit(X) ### km model is just used to find optimum k value, 
    Sum_of_squared_distances.append(km.inertia_)### inertia is the sum of squared distances
```

```python
### Plotting WSS plot
plt.figure(figsize=(7,4))
plt.plot(K, Sum_of_squared_distances, 'bx-') ###  'bx-' is used to show both line and the scatter points.
plt.xlabel('k')
plt.ylabel('Sum_of_squared_distances')
plt.title('Elbow Method For Optimal k')
plt.show()
```

From the above WSS plot, we consider k=4 as the optimum number of clusters

##### Problem 6.4 
Prepare a separate Data frame to show which stocks belong to the same cluster 

```python
### Creating new K-Means model'kmeans'
## n_init is thr Number of times the k-means algorithm will be run with different centroid seedst to eliminate any possible mistakes
kmeans= KMeans(n_clusters=4, n_init=25,random_state=1234) ## random state is fixed to obtain same cluster no. labels everytime
kmeans.fit(X)
```

```python
y_kmeans=kmeans.predict(X) ### Alternatively we can also use 'kmeans.labels_' 
y_kmeans
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X[y_kmeans==0,0],X[y_kmeans==0,1], label='Cluster 0')
plt.scatter(X[y_kmeans==1,0],X[y_kmeans==1,1], label='Cluster 1')
plt.scatter(X[y_kmeans==2,0],X[y_kmeans==2,1], label='Cluster 2')
plt.scatter(X[y_kmeans==3,0],X[y_kmeans==3,1], label='Cluster 3')

## plotting centroids
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=200, alpha=0.3)

plt.xlabel('Annual Returns')
plt.ylabel('Volatility')
plt.title('CLUSTERING')
plt.legend()
```

```python
### Adding the Cluster number to the original dataframe
Thirty_data['Cluster Number']=y_kmeans
Thirty_data
```

```python
### Plot to show which stocks belongs to same cluster
plt.figure(figsize=(17,8))
plt.scatter(Thirty_data['Annual Returns'],Thirty_data['Volatility'], c=Thirty_data['Cluster Number'], s=150, cmap='viridis')

plt.xlabel('Annual Returns')
plt.ylabel('Volatility')
plt.title('CLUSTER PLOT')


##### Adding label the data points 
n=Thirty_data.index ## Names of the companies

for i, txt in enumerate(n):
    plt.annotate(txt, (Thirty_data['Annual Returns'][i], Thirty_data['Volatility'][i]))
```
