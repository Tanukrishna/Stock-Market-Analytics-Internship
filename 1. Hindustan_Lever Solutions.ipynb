##### Problem 1.5
####### Write a function to calculate the average price over the last N days of the stock price data where N is a user defined parameter. Write a second function to calculate the profit/loss percentage over the last N days.
Calculate the average price AND the profit/loss percentages over the course of last -
1 week, 2 weeks, 1 month, 3 months, 6 months and 1 year.

```python
def AVG_price(N):
    return (stock_data['Average Price'].tail(N).sum())/N
print("Average prices for last N days :")
print("For last 1 week",AVG_price(5))
print("For last 2 weeks",AVG_price(10))
print("For last 1 month",AVG_price(20))
print("For last 3 months",AVG_price(60))
print("For last 6 months",AVG_price(120))
print("For last 1 year",AVG_price(240))
```

```python
def profit_loss(N):
    diff = stock_data['Close Price'].tail(N).iloc[N-1] - stock_data['Close Price'].tail(N).iloc[0]
    if diff < 0 :
        loss = abs(diff)
        loss_percent = (loss/stock_data['Close Price'].tail(N).iloc[N-1])*100
        return loss_percent
    if diff > 0 :
        profit = diff
        profit_percent = (profit/stock_data['Close Price'].tail(N).iloc[N-1])*100
        return profit_percent
print("Loss/Profit percentage for last N days are as follows:")
print("Last 1 week",profit_loss(5))
print("Last 2 weeks",profit_loss(10))
print("Last 1 month",profit_loss(20))
print("Last 3 months",profit_loss(60))
print("Last 6 months",profit_loss(120))
print("Last 1 year",profit_loss(240))
```

##### Problem 1.7
####### Add another column 'Trend' whose values are:
'Slight or No change' for 'Day_Perc_Change' in between -0.5 and 0.5
'Slight positive' for 'Day_Perc_Change' in between 0.5 and 1
'Slight negative' for 'Day_Perc_Change' in between -0.5 and -1
'Positive' for 'Day_Perc_Change' in between 1 and 3
'Negative' for 'Day_Perc_Change' in between -1 and -3
'Among top gainers' for 'Day_Perc_Change' in between 3 and 7
'Among top losers' for 'Day_Perc_Change' in between -3 and -7
'Bull run' for 'Day_Perc_Change' >7
'Bear drop' for 'Day_Perc_Change' <-7

```python
conditions = [
    (stock_data['Day_Perc_Change'] >= -0.5) & (stock_data['Day_Perc_Change'] <= 0.5),
    (stock_data['Day_Perc_Change'] >= 0.5) & (stock_data['Day_Perc_Change'] <= 1),
    (stock_data['Day_Perc_Change'] <= -0.5) & (stock_data['Day_Perc_Change'] >= -1),
    (stock_data['Day_Perc_Change'] >= 1) & (stock_data['Day_Perc_Change'] <= 3),
    (stock_data['Day_Perc_Change'] <= -1) & (stock_data['Day_Perc_Change'] >= -3),
    (stock_data['Day_Perc_Change'] >= 3) & (stock_data['Day_Perc_Change'] <= 7),
    (stock_data['Day_Perc_Change'] <= -3) & (stock_data['Day_Perc_Change'] >= -7),
    (stock_data['Day_Perc_Change'] > 7),
    (stock_data['Day_Perc_Change'] < -7)]
    
choices = ['Slight or No change','Slight positive', 'Slight negative','Positive', 'Negative','Among top gainers','Among top gainers','Bull run','Bear drop']

stock_data['Trend'] = np.select(conditions, choices, default='null')
stock_data
```
